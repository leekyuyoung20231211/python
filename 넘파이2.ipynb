조건식 표현

```python
ary1 = np.random.randint(0,100,size = (3,3))
ary2 = np.random.randint(0,100,size = (3,3))
cond = np.random.choice([ True,False],size = (3,3))
ary1,ary2,cond
```

```python
np.where(cond,ary1,ary2)   # 조건, 값1, 값2   조건에 맞으면 값1을 선택하고 그렇지 않으면 값2
```

축정보 axis 의 기능은
  - min() max() sum() 처럼 개별 요소에대해서 적용할때는
    - axis = 0이면 세로, 1이면 가로의 의미
  - 개별 요소가 아니라 행끼리 또는 컬럼끼리 적용할때는
    - axis = 0이면 행, 1이면 열
    - axis = 0이면 행전체, 1이면 행의 각 요소

```python
ary = np.array([
    [1,2,3,1],
    [4,5,6,4],
    [1,2,3,1]
])
np.unique(ary,axis=1)
```
