숫자 맞추기 게임
  - 컴퓨터가 임의의 숫자를 생성  random.randint(1,100) 1 ~ 99 임의의 자연수
  - 사용자로부터 값을 입력받아서  int(input())
  - 비교 if 구문
  - 사용자가 크면 크다, 작다 출력하고 맞추면 정답출력하고 종료

```python
import random
guess_number =  random.randint(1,100)  # computer
# print(guess_number)

while(guess_number != user_input):
  user_input =  int(input('1~100사이의 숫자를 입력하세요'))

  if user_input > guess_number:
    print('사용자의 값이 큽니다.')
  elif user_input < guess_number:
    print('사용자의 값이 작습니다.')
  else:
    print('정답입니다.')
```

```python
# 일차원 리스트는 과목별 점수
# 학생의 점수는 세로로 구해야 한다.
scores = [
    [49,80,20,100,80]
    ,[43,60,85,30,90]
    ,[49,82,48,50,100]
]
```

```python
total = [0,0,0,0,0]
for i in range(3):
  for j in range(5):
    total[j] += scores[i][j]

# [141, 222, 153, 180, 270]
student_mean = []
for i in total:
  student_mean.append(i/3)

print('각 학생별 총점 : ',total)
print('각 학생별 평균 : ',student_mean)
```

```python
import datetime
import pytz
def showCurrentTime():
  k_timezone =  pytz.timezone('Asia/Seoul')
  current_time = datetime.datetime.now()
  current_time = current_time.astimezone(k_timezone)
  current_time = current_time.strftime("%Y-%m-%d %H:%M:%S %Z")
  return current_time
```

```python
print('현재시간', showCurrentTime() )
```

함수는
  - 코드의 재 사용을 위해서..
  - 로직의 단순화, 명령어들을 설명할수 있음
  - 단순명료해 지기 때문에 실수를 줄여줌

```python
def factorial(n):
  if n == 1:
    return 1
  return n*factorial(n-1)
```

```python
a = factorial(  5  )
print(a)
```
