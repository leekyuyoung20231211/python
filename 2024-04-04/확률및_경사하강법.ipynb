인공신경망에서 "로짓"은 일반적으로 이진 분류 문제에서 사용되는 활성화 함수인 시그모이드 함수의 역함수를 나타냅니다. 시그모이드 함수는 입력 값을 0과 1 사이의 값으로 변환하는 함수로, 이는 이진 분류 문제에서 확률 값을 나타내는 데 사용됩니다. 그런데 시그모이드 함수의 역함수는 입력으로부터 확률을 구한 후 다시 해당 확률을 로짓이라고 하는 값으로 변환하는 것입니다.

로짓 함수는 시그모이드 함수의 역함수로서, 다음과 같이 정의됩니다:

로짓 함수는 $ \text{로짓}(p) = \ln\left(\frac{p}{1-p}\right) $ 로 표현됩니다.


여기서 \(p\)는 0과 1 사이의 확률 값을 나타냅니다. 로짓 함수를 사용하면 확률 값을 로그-오즈(odds)로 변환할 수 있습니다. 로그-오즈는 로짓 함수를 통해 얻은 값이며, 이는 특정 사건이 발생할 확률과 발생하지 않을 확률의 비율을 의미합니다.

로짓 함수는 주로 로지스틱 회귀와 같은 이진 분류 모델에서 사용되며, 이진 분류 문제에서 확률 값을 계산할 때 유용합니다. 로짓 값은 보통 이진 분류 문제에서 선형 결합된 입력 특성에 대한 가중치와 편향을 학습하여 계산됩니다. 계산된 로짓 값은 확률로 변환되어, 예측 클래스 레이블을 결정하는 데 사용됩니다.

입력을 벡터 x = |x1, x2, x3 ,,, xn |, 뉴런의 가중치를 w = |w1,w2,w3,,, wn|

뉴런의 출력은 y=f(x dot w + b)
