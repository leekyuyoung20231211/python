라인그래프 - plot 은 어떤종류의 데이터를 표현하는데 적합할까?

라인 그래프 (plot)는 시간에 따른 데이터의 변화를 시각적으로 나타내는 데 적합합니다. 특히 다음과 같은 상황에서 라인 그래프가 효과적입니다:

- 시계열 데이터 표현: 주가 변동, 기온 추이, 판매 추이 등과 같이 시간에 따라 변하는 데이터를 시각화할 때 라인 그래프가 자주 사용됩니다. 시간이 X축에 위치하고 해당 시점에서의 데이터 값이 Y축에 위치합니다.

- 연속적인 데이터 표현: 라인 그래프는 데이터의 연속적인 특성을 잘 나타내므로, 데이터가 연속적으로 변하는 경우에 사용됩니다. 예를 들어, 실험에서 시간이 흐름에 따라 어떤 물질의 농도가 어떻게 변하는지를 보여줄 때 사용할 수 있습니다.

- 함수 그래프 표현: 수학적인 함수의 그래프를 표현할 때도 라인 그래프가 유용합니다. 함수의 입력 값을 X축에 놓고 해당 값에서 함수가 반환하는 결과를 Y축에 놓아 표현할 수 있습니다.

- 트렌드 및 패턴 분석: 라인 그래프는 데이터의 추세, 주기성, 패턴 등을 시각적으로 확인하기에 용이합니다. 특히 시계열 데이터에서 어떤 추세가 있는지 파악하는 데 도움이 됩니다.

- 다중 라인 그래프: 여러 데이터셋을 동시에 비교하고 싶을 때 여러 개의 라인을 하나의 그래프에 표현할 수 있습니다. 각 라인의 색상, 선 종류 등을 조절하여 데이터 간 비교를 용이하게 할 수 있습니다.

라인 그래프는 데이터의 추세와 변화를 시각적으로 이해하는 데 도움이 되며, 데이터셋 간의 비교와 관계를 파악하는 데도 유용합니다.

누적막대 그래프(stacked bar graph)

```python
data = np.array([
    [5,25,50,20],
    [4,23,51,17],
    [6,22,52,19],
    [1,2,3,4]
])
color_list = ['b','g','r','y']
data_label = ['A','B','C','D']
x = np.arange(data.shape[1])
```

```python
plt.bar(x,data[:,0],bottom = [0,0,0,0])
plt.bar(x,data[:,1],bottom = [5,4,6,1])
plt.bar(x,data[:,2],bottom = [5+25,4+23,6+22,1+2])
plt.bar(x,data[:,3],bottom = [5+25+50,4+23+51,6+22+52,1+2+3])
```

```python
for i in range(4):
  plt.bar(x,data[:,i],bottom = np.sum(data[:,:i],axis=1),label= data_label[i],color=color_list[i])
plt.legend()
plt.show()
```

```python
g = sns.FacetGrid(data=tips, col = 'time', row='sex')
# 매개변수 col, row에 범주형 데이터를 넣으면 데이터 종류만큼 MxN의 그래프를 생성
```
