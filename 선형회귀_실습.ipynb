```python
import numpy as np
```

```python
np.random.seed(0)
x = 2*np.random.rand(100,1)  # 0~1사이의 데이터 중에 균일하게  분포된 데이터  100 x 1
y = 3*x+1+np.random.rand(100,1)
```

```python
# 시각화
# 데이터 분할
# 선형모델 학습
# 기울기(결정계수 R^2 , 회귀계수), 절편
# 예측
# 테스트데이터(scatter)와 예측값(plot)을 시각화
```

```python
# 시각화
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.show()
```

```python
# 데이터 분할
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test =  train_test_split(x,y,random_state=50)
```

```python
# 선형모델 학습
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
# 기울기(결정계수 R^2 , 회귀계수), 절편
print(f"기울기(결정계수 R^2 , 회귀계수) : {lr.coef_}")
print(f"절편 : {lr.intercept_}")
```

```python
# 예측
prediction = lr.predict(X_test)
prediction
print(f"테스트 데이터 스코어 : {lr.score(X_test,y_test)}")
```

```python
# 테스트데이터(scatter)와 예측값(scatter)을 시각화
import matplotlib.pyplot as plt
plt.scatter(X_test,y_test,label = 'test')
plt.scatter(X_test,prediction,label = 'prediction')
plt.legend()
plt.show()
```
