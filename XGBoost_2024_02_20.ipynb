교차검증
  - 선형 회귀 교차 검증

```python
from sklearn.model_selection import cross_val_score, cross_validate
model = LinearRegression()
# 10 폴드 교차검증으로 평균제곱오차
# neg_mean_squared_error 평균제곱오차에 음수로 표현
# 교차검증은 일반적으로 높은 값이 우수한 모델을 나타낸다 mse경우 값이 작을수록 모델이 더 좋다
# 음수를  사용해서 mse의 값이 커지면 더 나은 모델로 간주할수 있도록 하기위해서
scores = cross_val_score(model,X,y,scoring = 'neg_mean_squared_error', cv=10)
rmse = np.sqrt(-scores)
print(f"회귀 rmse : {rmse}")
print(f"rmse 평균 : {rmse.mean()}")
```

```python
scores = cross_validate(model,X,y,scoring = 'neg_root_mean_squared_error', cv=10)
-np.mean(scores['test_score'])
```

XGBoost 교차 검증

```python
model = XGBRegressor(objective='reg:squarederror')
# 10폴드
scores = cross_val_score(model,X,y,scoring='neg_mean_squared_error',cv=10)
rmse = np.sqrt(-scores)
print(f"회귀  rmse : {np.round(rmse,2)}")
print(f"rmse 평균 : {rmse.mean()}")
```

warm_start : 모델이 이전호출에서 학습한 모델에 추가적인 결정트리를 학습할수 있도록 한다
  - 반복학습에 사용
  - 대용량 데이터를 분할해서 학습할때

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()  # dark모드
oob_scores= []
df = pd.read_csv('https://raw.githubusercontent.com/leekyuyoung20231211/python/main/data/census_cleaned.csv')
rf = RandomForestClassifier(n_estimators=50, n_jobs=-1, random_state=2, oob_score=True,warm_start=True)
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
rf.fit(X,y)
oob_scores.append(rf.oob_score_)
est = 50
estimators = [est]
for i in range(9):
  est += 50
  estimators.append(est)
  rf.set_params(n_estimators=est)
  rf.fit(X,y)
  oob_scores.append(rf.oob_score_)

plt.figure(figsize=(15,7))
plt.xlabel("number of trees")
plt.ylabel('oob_score_')
plt.plot(estimators, oob_scores)
plt.show()
```

랜덤 포레스트 성능 높이기 - 사계연구


```python
df = pd.read_csv('https://raw.githubusercontent.com/leekyuyoung20231211/python/main/data/bike_rentals_cleaned.csv')
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=2)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(warm_start=True,n_jobs=-1,random_state=2)
# 시작 트리수
est = 10
rmse_scores = []
estimators = []
for i in range(21):
  rf.set_params(n_estimators = est)
  rf.fit(X_train,y_train)
  mse = mean_squared_error(y_test,rf.predict(X_test))
  rmse = np.sqrt(mse)
  rmse_scores.append(rmse)
  estimators.append(est)
  est += 25

plt.figure(figsize=(15,7))
plt.xlabel("number of trees")
plt.ylabel('rmse')
plt.plot(estimators, rmse_scores)
plt.show()
```
