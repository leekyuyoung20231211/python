```python
# 결측치 유무를 조사하고 처리방안 고민 - 비율 - 결측치가 있는 항목..
df.isnull().mean()[df.isnull().mean() > 0]
```

```python
# 불필요한 컬럼 제거
# Unnamed: 19  컬럼 삭제 - 드랍
# 도로명 또는 지번 둘중에 하나만 있는 데이터 대체 - 결측치
# 위경도가 국내 범위를 벗어나는 데이터는 제외 - 이상치
```

```python
df = df.drop(columns=['Unnamed: 19'])
df.head(1)
```

```python
# 소재지도로명주소 결측치를 소재지지번주소로 채운다
df['소재지도로명주소'] = df['소재지도로명주소'].fillna(df['소재지지번주소'])
df['소재지도로명주소'].isnull().sum()
```

주어진 피처(컬럼) 이 아닌 새롭게 만든는 컬럼
  - 파생변수
  - 주소에서 시도 구군 파생변수 생성

```python
df['시도'] = df['소재지도로명주소'].apply(lambda x : x.split()[0])
df['구군'] = df['소재지도로명주소'].apply(lambda x : x.split()[1])
```

```python
# 시도별 데이터 빈도수 구해보기
df['시도'].value_counts()
```

```python
df['시도'] = df['시도'].replace('강원','강원도')
df['시도'].value_counts()
```

```python
# 위경도의 이상치 제거  우리나라 범위를 벗어난 경우
# 경도, 위도로 산점도를 그려보면... 알수 있음
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(data=df, x='경도', y='위도')
plt.show()
```

```python
import numpy as np
print(True & False, True and False)
print(np.array([True,True,False]) & np.array([True,True,False]))
print(np.array([True,True,False]).any() and np.array([True,True,False]).any())
```

```python
# and  연산자는 좌우에 리스트가 올수 없다..   리스트끼리의 연산은 &를 이용
```

```python
df_park = df[(df['경도'] < 132) & (df['위도'] > 32)]
# (df['경도'] < 132) and (df['위도'] > 32)
sns.scatterplot(data=df_park, x='경도', y='위도')
plt.show()
```

```python
temp = re.compile('[가-힣]+').findall(gym)
temp = [name for name in temp if (len(name) > 1) & (name not in  ['운동기구','개소']) ]
from collections import Counter
temp = dict(Counter(temp))
sorted(temp.items(), key=lambda x: x[1],reverse=True)[:10]
```
