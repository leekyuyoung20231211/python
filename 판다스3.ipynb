데이터 병합
  - inner join : 교집합처럼 키가 같은 데이터만 병합
  - jull join : 합집합처럼 키가 같은 데이터는 병합하고 키가 다르면 각각표시
  - left join : 왼쪽을 기준으로(전부표시) 오른쪽은 키가 같은 데이터만 병합
    - right join 은 반대개념

```python
import pandas as pd
import numpy as np
```

```python
raw_data = {
'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
'test_score': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
```

```python
df_left = pd.DataFrame(raw_data)
```

```python
raw_data2 = {
            'subject_id': ['4', '5', '6', '7', '8'],
            'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
            'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']
            }
df_right = pd.DataFrame(raw_data2)
```

```python
df_left
```

```python
df_right
```

```python
pd.merge(left=df_left, right=df_right,how='inner', on='subject_id' )  # default inner
```

```python
# left
pd.merge(left=df_left, right=df_right,how='left', on='subject_id' )  # default inner
```

```python
# right
pd.merge(left=df_left, right=df_right,how='right', on='subject_id' )  # default inner
```

```python
# outer
pd.merge(left=df_left, right=df_right,how='outer', on='subject_id' )  # default inner
```

join에 참여하는 데이터프레임의 기준 컬럼명이 서로 다르면
  - on 대신에 left_on, right_on 매개변수를 사용한다
  - pd.merge(left=df_left, right=df_right,how='outer', left_on='id', right_on='no' )  # default inner

인덱스를 키 값으로 사용 할 수 있음

```python
df_left.index = df_left['subject_id']
df_right.index = df_right['subject_id']
```

```python
df_left.drop('subject_id',axis=1,inplace=True)
df_right.drop('subject_id',axis=1,inplace=True)
```

```python
pd.merge(df_left,df_right,on='subject_id')
```

```python
# 함수제작
def makeStore(file_path):
  cols = ['시도명','시군구명','행정동명','법정동명','상호명','상권업종대분류명','상권업종중분류명','상권업종소분류명','표준산업분류명']
  df = pd.read_csv( file_path ,low_memory=False)
  return df.loc[:,cols ]
```

```python
# 1차 표준산업분류명 == '커피 전문점'  해당하는 데이터를 전부 추출
# 2차 커피 전문점의 종류(이디아., 스타벅스...) 를 추출
# 3차 시도별 커피 전문점의 개수
# 4차 시도별 전문점의 종류 개수
temp = df[ df['표준산업분류명'] == '커피 전문점' ]
temp = temp['상호명'].value_counts()
temp = temp[temp.values >= 5]
temp = temp.sort_values(ascending=False)
store_top_10 = temp[:10]
```

```python
# 3차 시도별 커피 전문점의 개수
temp = df[df['표준산업분류명'] == '커피 전문점']
temp = temp.groupby('시도명')['표준산업분류명'].count().sort_values(ascending=False).to_frame()
temp.columns = ['점포수']
temp
```
