```python
import requests
import xmltodict
import json
import pandas as pd

def getDataFromAPI(url, params={}):
  '''getDataFromAPI(url, params={})
  parameters:
    url = 'http://openapi.tour.go.kr/openapi/service/EdrcntTourismStatsService/getEdrcntTourismStatsList'
    params ={
          ,'YM' : '202001'
          ,'NAT_CD' : '112'
          ,'ED_CD' : 'E'
          }'''
  params['serviceKey'] = 'U5i63xpIWM48raIRQBYpUXbA/wKd0iO6n/n1+JISyoEW1gWcEiz2No2fHPeid5TZoQN0HV85WyGv6LPwaQ8n4w=='
  response = requests.get(url, params=params)
  result =  xmltodict.parse(response.text)
  result = result['response']['body']['items']['item']
  df = pd.DataFrame(result)
  df.columns = [ '출/입국 구분','출/입국 구분코드','국가코드','국가','출/입국자 수','결과값 연번','년도' ]
  # df.drop('결과값 연번', axis=1)
  df = df.drop(columns=['출/입국 구분코드','국가코드', '결과값 연번'])
  return df
```

```python
url = 'http://openapi.tour.go.kr/openapi/service/EdrcntTourismStatsService/getEdrcntTourismStatsList'
param = {'YM' : '202311'}
result = getDataFromAPI(url,param)
result

# '202101','202102' ~~ '202311'
```

최근 3년간 우리나라를 방문한 top 5 국가를 찾고 방문자수를 출력

```python
# 함수를 이용해서 36개월치 DataFrame을 리스트에 담는다  try 구문을 이용해서 제공되는 정보까지 추출한다
# concat을 이용해서 합친다
# 인덱스가 뒤죽박죽이므로 reset_index(drop=True)
# 국가별로 groupby해서 방문자수를 sum한다
# 내림차순으로 정렬한다
# 정렬된상태에서 5개만 추출한다. --> top5
```

```python
import itertools

start_year = 2021
end_year = 2023

start_month = 1
end_month = 12

date_codes = [f'{year}{month:02}' for year, month in itertools.product(range(start_year, end_year + 1), range(start_month, end_month + 1))]

# date_codes = []
# for year in range(2021,2024):
#   for month in range(1,13):
#     date_codes.append(f'{year}{month:02}')
date_codes = date_codes[ :-1 ]
print(date_codes)
```

```python
from tqdm import tqdm
url = 'http://openapi.tour.go.kr/openapi/service/EdrcntTourismStatsService/getEdrcntTourismStatsList'

result = []
for yearmonth in tqdm(date_codes):
  param = {'YM' : yearmonth}
  result.append(getDataFromAPI(url,param))
```

```python
total_df = pd.concat(result)
total_df = total_df.reset_index(drop=True)
total_df
```

```python
total_df['출/입국자 수'] = total_df['출/입국자 수'].astype('int')
```

```python
total_df2 =  total_df.groupby('국가')['출/입국자 수'].sum()
total_df2
```

```python
total_df2 = total_df2.drop(index='한  국')
```

```python
total_df2.sort_values(ascending=False)[:5]
```
