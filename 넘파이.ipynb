- 위의 코드를 참고해서 3x3 크기의 넘파이 배열을 생성
- 그리고 모든 값을 2개로 증가시킨다
- 생성한 넘파이 배열의 평균값을 계산

```python
SIZE = 3
temp = np.random.randint(0,100,size=(SIZE,SIZE))
print(temp)
print('-'*30)
temp *= 2
print(temp)
```

```python
# 평균
total = 0
for i in range(SIZE):
  for j in range(SIZE):
    total += temp[i][j]
print('평균',total / SIZE*SIZE)
```

```python
# 평균
total = 0
for i in range(SIZE):
    total += sum(temp[i])
print('평균',total / SIZE*SIZE)
```

```python
# 파이썬 리스트에서 슬라이싱은 해당결과를 복사해서 새로운 데이로 만든다
ary_sub1 = python_ary[ 3: 7]
print(ary_sub1)
ary_sub1[0] = 100; ary_sub1[1] = 100; ary_sub1[2] = 100; ary_sub1[3] = 100
print(ary_sub1)
print(python_ary)
```

```python
# 넘파이에서 슬라이싱은 새로운게 아니라 기존 데이터의 메모리를 공유한다..
print(ary)
ary_sub2 = ary[ 3: 7]
# ary_sub2[0] = 100; ary_sub2[1] = 100; ary_sub2[2] = 100; ary_sub2[3] = 100
ary_sub2[ : ] = 100   # 넘파이 브로드 캐스팅
print(ary_sub2)
print(ary)
```

```python
# 리스트의  슬라이싱 --> 데이터를 분리해서 사용 copy
#넘파이의 슬라이싱 --> 공유  copy가 아님
# 넘파이의 copy???
import numpy as np
ary = np.arange(10)
print(ary)
# ary_sub3 = ary[3:7].copy()
ary_sub3 = ary[3:7]
print(ary_sub3)
ary_sub3[:] = 100
print('-'*20)
print(ary_sub3)
print(ary)
```
