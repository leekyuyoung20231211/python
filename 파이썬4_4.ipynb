```python
# pip install beautifulsoup4  --> 로컬환경
from bs4 import BeautifulSoup
# 한글이 포함되어 있으나 깨지면 cp949 or ms949
with open("books.xml",'r',encoding='utf-8') as f :
  book_xml =  f.read()
```

```python
print(book_xml)
```

```python
soup =  BeautifulSoup(book_xml,"lxml")  # lxml파서를 사용해서 데이터를 분석
# author 태그의 모든 요소를 추출
for book in soup.find_all('author'):
  print(book)
  print(book.get_text())
```

```python
with open('/content/US08621662-20140107.XML') as f:
  xml_ = f.read()
soup = BeautifulSoup(xml_,'lxml') # xml 파서
# 태그 정보 추출
invention_title_tags = soup.find_all('invention-title')
for tag in invention_title_tags:
  print(tag.get_text())
```

```python
url = 'https://news.naver.com/'
import urllib.request
html_ = urllib.request.urlopen(url)
xml_ = html_.read()
# print(xml_)
soup = BeautifulSoup(xml_,'html.parser')  #html 파서
# 태그 정보 추출
invention_title_tags = soup.find_all('div', class_='cjs_news_tw')
invention_title_tags
for tag in invention_title_tags:
  print(tag.get_text())

```
