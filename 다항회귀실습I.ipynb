```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# 데이터 생성
np.random.seed(0)
X = 2*np.random.rand(100,1) - 1
y = 2*X**2 - 3*X +1 + np.random.rand(100,1)
```

```python
# 시각화
plt.scatter(X,y)
plt.show()
```

```python
# 선형회귀(다항)
model = LinearRegression()
model.fit(X,y)
model.coef_, model.intercept_
```

```python
prediction = model.predict(X)
# 시각화
plt.scatter(X,y,label='real data')
plt.scatter(X,prediction, label='prediction data')
plt.legend()
plt.show()
```

```python
# 학습데이터가 2차원이여야 함.
X_poly = np.column_stack([X**2,X])
model_poly = LinearRegression()
model_poly.fit(X_poly,y)
prediction_poly = model_poly.predict(X_poly)
# 시각화
plt.scatter(X,y,label = 'original_data')
plt.scatter(X,prediction_poly,label='poly prediction')
plt.show()
```
