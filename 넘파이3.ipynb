넘파이 배열의 데이터는 동일타입을 가진다

  - 정수, 실수 -->  실수
  - 객체가 포함되어 있으면 --> 객체 Object
  - 객체는 없고 문자열이 있으면 --> 문자열

```python
tensor_rank2 = [
    [1,2,5,8],[1,2,5,8],[1,2,5,8]
]
tensor_rank2 = np.array(tensor_rank2)
tensor_rank2, np.sum(tensor_rank2,axis=0)
```

텐서의 모양을 변경하기
  - reshape

```python
# 4 x 2 넘파이 배열을 생성

# 생성된  4 x 2 --> 2 x 4

#  2 x 4 -> 2 x 2 x 2

#  마지막은 1차원으로   넘파이배열객체.flatten()  --> reshape(-1)
```

```python
ary1 = np.arange(8)
print(ary1)
ary1 = ary1.reshape(4,-1)
print(ary1)
ary1 = ary1.reshape(2,-1)
print(ary1)
ary1 = ary1.reshape(2,2,-1)
print(ary1)
ary1 = ary1.flatten()
print(ary1)
```

연결함수
  - concat : 두 객체간의 결합을 지원하는 함수
  - vstack : 배열을 수직으로 붙임
  - hstack : 배열을 수평으로 붙임

image.png

```python
v1 = np.array([1,2,5])
v2 = np.array([2,3,4])

v3 = np.array([1,2,3]).reshape(-1,1)
v4 = np.array([2,3,4]).reshape(-1,1)

v1,v2,v3,v4
```

```python
np.vstack( (v1,v2) ), np.hstack( (v3,v4) )
```

```python
# V1, V2를 concat함수를 써서 vstack,hstack 의 내용으로 출력
```

```python
# 2차원 배열에 대한 축을 기준으로 하는 결합
np_v1 = np.array([v1])
np_v2 = np.array([v2])
np.concatenate( (np_v1,np_v2 ), axis = 0 ), np.concatenate( (np_v1.T,  np_v2.T ), axis = 1 )
```

all, any
  - all : 배열의 모든값이 참일때 참이고 그렇지 않으면 거짓
    - and 조건을 모든 요소에 적용
  - any : 배열의 모든값이 거짓일때만 거짓이고 그렇지 않으면 함
    - or 조건을 모든 요소에 적용

```python
(x > 5 ).all()
```

```python
(x > 5 ).any()
```
