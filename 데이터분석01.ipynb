```python
import pandas as pd
# 지역별 평당분양가격의 평균
pd.DataFrame( df.groupby(['지역명'])['평당분양가격'].mean())
```

```python
pd.pivot_table(df,index=['지역명'], values=['평당분양가격'], aggfunc='mean')
```

```python
# 지역명,전용면적으로 평당분양가격의 평균
# df.groupby(['전용면적','지역명'])['평당분양가격'].mean().unstack().round()
df.pivot_table(index=['전용면적'], columns=['지역명'],values=['평당분양가격'],aggfunc='mean').round()
```

```python
# 전용면적으로 분양가격의 평균을 구하고 막대그래프로 시각화
temp = df.groupby(['전용면적'])['분양가격'].mean().sort_values(ascending=False)
plt.bar(x = temp.index, height=temp.values)
plt.show()
```

- 표준편차 : 분산의 양의 제곱근, 작을수록 평균값에서 각값들(변량)의 거리가 가깝다
- 편차(오차) : 관측값(실제값)에서 평균이나 중앙값을 뺀값
- 분산 : 편차를 제곱한 값들의 평균( 편차를 모두더하면 0이 나온다 그래서 제곱)
- 표준편차 : 분산의 제곱근(루트) ,제곱해서 커진 값을 원래상태로 돌린다.

```python
# 표준편차-> 분산에 root를 씌운다
import numpy as np
np.std(df2['위도'])
```

```python
# df2['위도'].var()  # 분산
np.sqrt(df2['위도'].var())
```
